```python
df.drop(columns='description', inplace=True)
```

```python
df.head()
```

```python
df.drop(columns='cast', inplace=True)
```

```python
df.head()
```

```python
df['release_year'].value_counts().plot(legend=True, figsize=(10, 3))
plt.title('year in which most no. of movies and tvshows released on amazon',fontdict={"fontweight":"bold"})
plt.xlabel('year')
plt.ylabel('no. of movies released')
```

```python
df['rating'].isnull().value_counts()
```

```python
df['rating'].value_counts().plot(kind='bar')
plt.title('content availabe on amazon according to the age category')
plt.xlabel('age category')
plt.ylabel('no. of shows & movies available')
```

```python
sns.histplot(y='rating', hue='type', data=df)
plt.title('tvshows and movies available for different categories')
```

```python
genre_list = df['listed_in']
genre_list
```

```python
genre = []

for gen in genre_list:
    x = gen.split(',')
    genre.append(x[0])

genre
```

```python
df.drop(columns='listed_in', inplace=True)
df['genre'] = genre
df.head()
```

```python
df['genre'].value_counts().plot(kind='pie', legend=True)
plt.title('different genres available')
```

```python
sns.histplot(y='rating', hue='genre', data=df, alpha=1)
plt.title('different genres available as per the rating category')
```

```python
sns.histplot(y='genre', hue='type', data=df)
plt.title('tv shows and movies available according to different genre')
```

```python
df['director'].unique()
```

```python
df['director'].value_counts()[0:10].plot(kind='bar')
plt.title('top directors to produce the movie')
```

```python
df.dropna(subset =['country'], inplace=True)
```

```python
df['country'].isnull().value_counts()
```

```python
df['country'].value_counts()[0:10].plot(kind='bar')
plt.title('top 10 countries to release the movies on amazon')
```

```python
indian_content = df[df['country']=='India'].copy()
indian_content
```

```python
sns.histplot(x='genre', hue='type', data=indian_content)
plt.title('shows and movies released by india')
plt.xticks(rotation='vertical')
```

```python
indian_content['director'].value_counts()[0:10].plot(kind='bar')
plt.title('top directors of india')
```
